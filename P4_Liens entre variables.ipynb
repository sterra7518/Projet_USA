# 2 - Test statistiques

En statistiques, un test, ou test d'hypothèse, est une procédure de décision entre deux hypothèses. Il s'agit d'une démarche consistant à rejeter ou à ne pas rejeter une hypothèse statistique, appelée hypothèse nulle, en fonction d'un échantillon de données.

Il s'agit de statistique inférentielle : à partir de calculs réalisés sur des données observées, on émet des conclusions sur la population, en leur rattachant des risques d'être erronées.

### 2-2 Test de Khi2


#### Tableau de contingence

```python
#Tableau de contingence
usa_crosstab = pd.crosstab(usa['PartUSA'],usa['Candidat'],margins = True)


usa_crosstab
```

#### Calcul du Khi-2

```python
from scipy.stats import chi2_contingency as chi2_contingency

khi2, pval , ddl , contingent_theorique = chi2_contingency(usa_crosstab)
```

```python
#Distance du Khi-2
khi2
```

```python
#La p-valeur
pval
```

*La p-valeur est de 24,1% qui est le seuil ou on peut rejetter ou non les hypothèses*

```python
#Dégrée de liberté
ddl
```

```python
#Effectif théorique

pd.DataFrame(contingent_theorique)
```

### 2-4 Test de la moyenne

```python
#Population moyenne des communes étudiés

round(np.mean(usa.population))
```

Les communes étudiés ont une moyenne d'environ 314 423 habitants.On pourrait faire un test statistique pour voir si on pouvait avoir une moyenne de 314 000 sur un test de seuil à 5%

- hypothèse nulle : H0 : x = 314 000 
- hypothèse alternative : H1 : x =! 314 000

```python
import statsmodels
import scipy
import math

pop = list(usa['population'])


scipy.stats.ttest_1samp(list(pop),popmean = 314000)
```

La p-valeur est de 99% donc on ne rejete pas l'hypothèse nulle H0 et donc on peut arrondir la moyenne a 314000 habitants.

### 2-2 Répartition éthnique

Les Etats-Unis est divisés en plusieurs groupes ethnique :
  - population blanche : premiers immigrés provenant de l'Europe au XVIIe siècle
  - population noire : population des déscendants des esclaves au XIXe siècle
  - population hispanique : les principaux immigrés d'aujourd'hui
  - population asiatique : immigrés des grendes villes
 


```python
#On calcule les proportions des communautés



#on va créer un fonction pour faire des proportions
pop = list(usa.POP10)

#population blanche
prop_blanche = list(usa.WHITEp/100)
pop_blanche = list()
for i in range(917):
    pop_blanche.append(round(pop[i]*prop_blanche[i]))
    
#population noire
prop_noir = list(usa.BLACKp/100)
pop_noir = list()
for i in range(917):
    pop_noir.append(round(pop[i]*prop_noir[i]))

# population hispanique
prop_hispa = list(usa.HISPAp/100)
pop_hispa = list()
for i in range(917):
    pop_hispa.append(round(pop[i]*prop_hispa[i]))

#population asiatique
prop_asia = list(usa.ASIANp/100)
pop_asia = list()
for i in range(917):
    pop_asia.append(round(pop[i]*prop_asia[i]))

# population autres
prop_other = list(usa.OTHERp/100)
pop_other = list()
for i in range(917):
    pop_other.append(round(pop[i]*prop_other[i]))

    
    
pop_etno = [sum(pop_blanche),sum(pop_noir),sum(pop_hispa),sum(pop_asia),sum(pop_other)]    
pop_etno    
```

#### Indice de Gini

```python
#On calcul l'indice de Gini

indice = []

#population blanche
dep = usa['WHITEp'].values
n = len(dep)
lorenz = np.cumsum(np.sort(dep)) / dep.sum()
lorenz = np.append([0],lorenz) # La courbe de Lorenz commence à 0
AUC = (lorenz.sum() -lorenz[-1]/2 -lorenz[0]/2)/n # Surface sous la courbe de Lorenz. Le premier segment (lorenz[0]) est à moitié en dessous de 0, on le coupe donc en 2, on fait de même pour le dernier segment lorenz[-1] qui est à moitié au dessus de 1.
S = 0.5 - AUC # surface entre la première bissectrice et le courbe de Lorenz
gini = round(2*S,2)
gini
indice.append(gini)


#population noire
dep = usa['BLACKp'].values
n = len(dep)
lorenz = np.cumsum(np.sort(dep)) / dep.sum()
lorenz = np.append([0],lorenz) # La courbe de Lorenz commence à 0
AUC = (lorenz.sum() -lorenz[-1]/2 -lorenz[0]/2)/n # Surface sous la courbe de Lorenz. Le premier segment (lorenz[0]) est à moitié en dessous de 0, on le coupe donc en 2, on fait de même pour le dernier segment lorenz[-1] qui est à moitié au dessus de 1.
S = 0.5 - AUC # surface entre la première bissectrice et le courbe de Lorenz
gini = round(2*S,2)
gini
indice.append(gini)


#population hispanique
dep = usa['HISPAp'].values
n = len(dep)
lorenz = np.cumsum(np.sort(dep)) / dep.sum()
lorenz = np.append([0],lorenz) # La courbe de Lorenz commence à 0
AUC = (lorenz.sum() -lorenz[-1]/2 -lorenz[0]/2)/n # Surface sous la courbe de Lorenz. Le premier segment (lorenz[0]) est à moitié en dessous de 0, on le coupe donc en 2, on fait de même pour le dernier segment lorenz[-1] qui est à moitié au dessus de 1.
S = 0.5 - AUC # surface entre la première bissectrice et le courbe de Lorenz
gini = round(2*S,2)
gini
indice.append(gini)



#population asiatoque
dep = usa['ASIANp'].values
n = len(dep)
lorenz = np.cumsum(np.sort(dep)) / dep.sum()
lorenz = np.append([0],lorenz) # La courbe de Lorenz commence à 0
AUC = (lorenz.sum() -lorenz[-1]/2 -lorenz[0]/2)/n # Surface sous la courbe de Lorenz. Le premier segment (lorenz[0]) est à moitié en dessous de 0, on le coupe donc en 2, on fait de même pour le dernier segment lorenz[-1] qui est à moitié au dessus de 1.
S = 0.5 - AUC # surface entre la première bissectrice et le courbe de Lorenz
gini = round(2*S,2)
gini
indice.append(gini)

#Transfomation en Data Frame
indice = pd.DataFrame(indice)
indice = indice.T
indice.rename(columns = {0: "blanche", 1: "noire", 2:"hispanique" ,3:"asiatique"},index = {0:"Indice de Gini"}, inplace = True)


indice
```

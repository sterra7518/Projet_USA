```python
#Importation des modules de Python 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Ouverture des fichiers 


```python
#Ouverture de notre jeux de donnÃ©es principale
usa =  pd.read_csv('usa2.csv', sep=',', encoding='ISO-8859-1')

#Suppresion de la 1er colonne suivante dÃ©rengeante
usa = usa.drop(['Unnamed: 0'],axis=1)

usa.head(10)
```

```python
usa.columns
```

# 1 - CrÃ©ation d'une nouvelle variable



## 1-1 Construction

```python
#SÃ©paration des donnÃ©es en 4 types de communes

population = usa.POP10

#conversion en une liste
population = list(population)

#Afficher la liste
population 
```

```python
#CrÃ©ation d'une nouvelle variable

type_ville = list()

for i in range(len(population)):
    
    if population[i] < 25000 :
        type_ville.append("Petite ville")
    
    elif population[i]> 25000 and population[i] < 100000:
        type_ville.append("Moyenne ville")
        
    elif population[i]> 100000 and population[i] < 1000000:
        type_ville.append("Grande ville")
    
    else:
        type_ville.append("MÃ©tropÃ´le")

```

```python
#Conversion en DataFrame
type_ville = pd.DataFrame(type_ville)

#modification du nom
type_ville = type_ville.rename(columns={0:"Type_ville"})

type_ville.head(10)
```

##### Fusion avec notre jeu de donnÃ©es


```python
cbssa = usa.cbssa
cbssa = pd.DataFrame(cbssa)

cbssa.head(10)
```

```python
ville = pd.concat([cbssa,type_ville], axis=1)

ville.head(10)
```

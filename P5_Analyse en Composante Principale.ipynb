```python
usa_acp = usa.drop(columns=["name", "state",'PartUSA',"NameState","Political","Candidat","BorderSea","Anglais_Off"],axis=1)
usa_acp.head(5)
```

#### Décomposition des données

```python
#on récupère tous les valeurs pour chaque ligne
X = usa_acp.values

#On récupère tous les individus qui sont en réalité les pays
names = usa_acp.index

# On récupère les colonnes dont on a besoin pour les utiliséq
features = usa_acp.columns
```

```python
# Importation des modules ACP par Scikit-Learn

from sklearn import decomposition
from sklearn import preprocessing
```

#### Matrice de central-réduite

```python
#Centrage et Réduction des données
sc=preprocessing.StandardScaler()

std_scale = sc.fit(X)
X_scaled = std_scale.transform(X)

X_scaled
```

```python
acp_usa = pd.DataFrame(X_scaled)

#Renommage des colonnes
acp_usa = acp_usa.rename(columns = {0:'population',
                             1:'pop_blanche',
                             2:'pop_noir',
                             3:'pop_hispanique',
                             4:'pop_asiatique',
                             5:'pop_autres',
                             6: '-18ans',
                             7:'18_24ans',
                             8:'25_44ans',
                             9:'45_64ans',
                             10:'+65ans',
                             11:'non-diplomés',
                             12 : 'lycée',
                             13:'bac',
                             14:'université',
                             15:'-20d',
                             16:'20_40d',
                             17:'40_60d',
                             18 : '60_80d',
                             19 : '+80d'})

acp_usa.to_csv('tableau_centre-reduite.csv',header = True)

acp_usa
```

#### Valeurs propres

```python
#Valeurs propres

pca_train = decomposition.PCA(n_components=20)
pca_train.fit(X_scaled)



print(pca_train)
```

```python
#Les part de variance expliqué par composante principale
print(pca_train.explained_variance_ratio_)
```

```python
#Les part de variance cumulés expliqué 
print(pca_train.explained_variance_ratio_.cumsum())
```

```python
#Valeur de l'inertie
val_propre = pd.DataFrame(pca_train.explained_variance_ratio_)
val_propre = val_propre.rename(columns = {0:'val_propre'})

#valeur de l'innertie en cumulé
val_propre2 = pd.DataFrame(pca_train.explained_variance_ratio_.cumsum())
val_propre2 = val_propre2.rename(columns = {0:'val_propre cumulé'})


#Rassembler les deux colonnes
val_propre3 = pd.concat([val_propre,val_propre2],axis=1)


#val_propre3 = val_propre3.rename(columns = {0:'val_propre',0:'val_propre cumulé'})

val_propre3
```

```python
val_propre3.head(10)['val_propre'].plot(kind='bar')

plt.xlabel("Valeurs propres")
plt.ylabel("Inertie")
plt.title("Valeurs propres")

plt.savefig("Valeurs propres.png")
plt.show() 
```

On a 45.6% de la variance totale récupéré

#### Vecteurs propres

```python
#Valeurs propres

pca_train = decomposition.PCA(n_components=2)
pca_train.fit(X_scaled)

#Vecteurs propres
pcs_train = pca_train.components_
variable = pd.DataFrame(pcs_train)



variable = variable.rename(columns = {0:'population',
                            1 : 'pop_blanche',
                            2 : 'pop_noire',
                            3 : 'pop_asia',
                            4 : 'pop_autres',
                            5 : 'pop_hispa',
                            6 : 'moins_18ans',
                            7 : 'de18_24ans',
                            8 : 'de25_44ans',
                            9 : 'de45_64ans',
                            10 : 'plus_65ans',
                            11 : 'no_diplome',
                            12 : 'lycee',
                            13 : 'bac',
                            14 : 'universite',
                            15 : 'moins_20d',
                            16 : 'de20_40d',
                            17 : 'de40_60d',
                            18 : 'de60_80d',
                            19 : 'plus_80d'})

variable = variable.rename(index = {0:'Richesse-Réussite',
                            1 : 'Diversité-Jeunesse'})


#Enregistrer en format CSV
variable.to_csv('var_acp.csv',header=True)


variable.T
```
